```python
##check given character is digit or not whithout using inbuilt func:

ch = input()
if len(ch)==1 and '0'<=ch<='9':
  print('is a digit or number')
else:
    print('is not a digit but an alphabet ')
```

```python
#check if given word is palindrme or not
word= input()
if word == word[::-1]:
 print("is a palindrome")
else:
  ("not a palindrome")
```

```python
#converts to binary, octal and hexadecimal
n = int(input())
print(bin(n))
print(oct(n))
print(hex(n))

```

```python
#check given integer is 3 digit int or not wihout using typecasting
num = int(input())
if (100 <= num <= 999) or (-999 <= num <= -100):
    print("It is a 3-digit integer")
else:
    print("It is not a 3-digit integer")
```
